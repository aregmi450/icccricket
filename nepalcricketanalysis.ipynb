```python
#importing modules
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset
df = pd.read_csv('international_results.csv')
df.head()
```

### Exploratory Data Analysis

Finding the total match results based on winning team

```python
df['Margin'] = df['Margin'].str.replace('-','No Result')
```

```python
# Select rows containing 'runs' in the 'win_margins' column
runs_rows = df[df['Margin'].str.contains('runs', case=False, na=False)]

# Convert the selected win margins to numeric values (remove 'runs' and convert to integers)
runs_rows['Margin'] = runs_rows['Margin'].str.replace(' runs', '').astype(int)

# Calculate the total win margins for runs
total_runs_win_margins = runs_rows['Margin'].value_counts().sum()

print("Total win margins in runs:", total_runs_win_margins)
```

```python
# Select rows containing 'runs' in the 'win_margins' column
wicket_rows = df[df['Margin'].str.contains('wickets', case=False, na=False)]

# Convert the selected win margins to numeric values (remove 'runs' and convert to integers)
# wicket_rows['Margin'] = wicket_rows['Margin'].str.replace(' wickets', '').astype(int)

# Calculate the total win margins for runs
total_wickets_win_margins = wicket_rows['Margin'].value_counts().sum()

print("Total win margins in wickets:", total_wickets_win_margins)
```

```python
# Select rows containing 'runs' in the 'win_margins' column
no_result = df[df['Margin'].str.contains('No Result', case=False, na=False)]

# Convert the selected win margins to numeric values (remove 'runs' and convert to integers)
# wicket_rows['Margin'] = wicket_rows['Margin'].str.replace(' wickets', '').astype(int)

# Calculate the total win margins for runs
total_no_results = no_result['Margin'].value_counts().sum()

print("Total matches with no result :", total_no_results)
```

```python
# Create data for the pie chart
data = [total_wickets_win_margins, total_runs_win_margins, total_no_results]
labels = ['Second Team Wins', 'First Team Wins', 'No Result']

# Create a pie chart using Seaborn
plt.figure(figsize=(4,5))
sns.set_palette("pastel")
plt.pie(data, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Win-Loss Distribution')
plt.show()
```

### Nepal Cricket Team Only Analysis 

Q.1 What is the total number of matches Nepalese team played ?

```python
# Getting the count of matches played by the country in team 1 and team 2 seperately.
t1=df['Team 1'].value_counts()
t2=df['Team 2'].value_counts()
```

```python
# Getting the total count of matches played by adding counts from team 1 and team 2.
final = t1+t2
final = final.reset_index()
final
```

Nepal team played 57 ODI Matches between 2015 and 2023.

Q.2. What is the count of matches won by Nepal ? 

```python
win = df.Winner.value_counts()
win = win.reset_index()
win
```

Nepal won 30 ODI Matches.

Q.3. What is the winning percentage of Nepal ? 

```python
#dropping null values for calculation
final.dropna(inplace= True)
final
```

```python
# Merging the data of total matches played and won by each country
final=final.merge(win)
final
```

```python
# Renaming the Columns and resetting the index for final dataframe.
final.rename(columns = {0:'Total Matches'}, inplace = True)
final.rename(columns = {'index':'Country'}, inplace = True)
final.set_index('Country',inplace=True)
final
```

```python
# Applying the winning percentage value by following equation.
final["Winning %"]=(final['Winner']/final['Total Matches'])*100
final["Winning %"]= final["Winning %"].round(2)
final
```

Nepal has a winning percentage of 52.63% after 57  total matches played.

```python
# Fetching the data of Team Nepal and converting it into DataFrame for further operation .
Nepal=final.loc['Nepal']
Nepal=pd.DataFrame(Nepal)

# Transposing the data and applying a variable to it Nepal1
Nepal1=Nepal.T
```

```python
# Applying a new column with the lost count of matches by team India
# Renaming the column Winner by Won
Nepal1['Lost']=Nepal1['Total Matches']-Nepal1['Winner']
Nepal1.rename(columns={"Winner":"Won"},inplace=True)
Nepal1
```

```python
# Calculate the sum of 'won' and 'lost' values
total_won = Nepal1['Won'].sum()
total_lost = Nepal1['Lost'].sum()

# Create data for the pie chart
data = [total_won, total_lost]
labels = ['Won', 'Lost']

# Create a pie chart using Seaborn
plt.figure(figsize=(4,5))
sns.set_palette("pastel")
plt.pie(data, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.title('Win-Loss Distribution')
plt.show()

```
